## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:**

    - Specific: Fokus pada tren pertumbuhan jumlah pesanan bulanan di platform Olist.
    - Measurable: Diukur melalui jumlah pesanan per bulan dan persentase pertumbuhan MoM.
    - Action-Oriented: Mengetahui pola musiman membantu perencanaan stok, kampanye marketing, dan alokasi sumber daya.
    - Relevant: Pertumbuhan jumlah pesanan adalah KPI utama bagi platform e-commerce.
    - Time-bound: Dibatasi pada periode 2017-2018.
- **Pertanyaan 2:**
"Kategori produk apa yang memiliki rata-rata skor review terendah (di bawah 4.0) dan volume penjualan signifikan (>100 transaksi) selama periode 2017-2018 sehingga perlu diprioritaskan untuk perbaikan kualitas produk atau layanan?"

    - Specific: Fokus pada kategori produk dengan performa review buruk namun volume tinggi.
    - Measurable: Diukur melalui rata-rata skor review dan jumlah transaksi per kategori.
    - Action-Oriented: Identifikasi kategori bermasalah agar tim bisa melakukan perbaikan kualitas.
    - Relevant: Kepuasan pelanggan (review score) langsung memengaruhi retensi dan reputasi.
    - Time-bound: Dibatasi pada periode 2017-2018.

## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
```

## Data Wrangling

### Gathering Data

#### Load df ...

```python
# ============================================================
# CELL: Gathering Data - Load semua DataFrame
# ============================================================
base_path = "data/"

customers_df = pd.read_csv(base_path + "customers_dataset.csv")
orders_df = pd.read_csv(base_path + "orders_dataset.csv")
order_items_df = pd.read_csv(base_path + "order_items_dataset.csv")
order_payments_df = pd.read_csv(base_path + "order_payments_dataset.csv")
order_reviews_df = pd.read_csv(base_path + "order_reviews_dataset.csv")
products_df = pd.read_csv(base_path + "products_dataset.csv")
sellers_df = pd.read_csv(base_path + "sellers_dataset.csv")
category_translation_df = pd.read_csv(base_path + "product_category_name_translation.csv")

print("=== Shape dari setiap dataset ===")
for name, df in [("customers", customers_df), ("orders", orders_df),
                 ("order_items", order_items_df), ("order_payments", order_payments_df),
                 ("order_reviews", order_reviews_df), ("products", products_df),
                 ("sellers", sellers_df), ("category_translation", category_translation_df)]:
    print(f"{name}: {df.shape}")

# ============================================================
# CELL: Gathering Data - Preview setiap DataFrame
# ============================================================
print("\n=== Orders DataFrame ===")
print(orders_df.head())
print("\n=== Order Items DataFrame ===")
print(order_items_df.head())
print("\n=== Products DataFrame ===")
print(products_df.head())
print("\n=== Order Reviews DataFrame ===")
print(order_reviews_df.head())
```

**Insight:** (Opsional)
- Dataset terdiri dari 8 tabel yang saling terhubung melalui order_id, product_id, customer_id, dan seller_id.
- orders_df memiliki ~99K baris, order_items_df ~112K baris, menunjukkan beberapa pesanan memiliki lebih dari 1 item.

### Assessing Data

#### Identifikasi Missing Values, Identifikasi Duplikasi, Cek Tipe Data

```python
# ============================================================
# CELL: Assessing Data - Identifikasi Missing Values
# ============================================================
print("=" * 60)
print("ASSESSING DATA")
print("=" * 60)

print("\n--- Missing Values per DataFrame ---")
for name, df in [("orders", orders_df), ("order_items", order_items_df),
                 ("products", products_df), ("order_reviews", order_reviews_df)]:
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if len(missing) > 0:
        print(f"\n{name}:")
        print(missing)
    else:
        print(f"\n{name}: Tidak ada missing values")

# ============================================================
# CELL: Assessing Data - Identifikasi Duplikasi
# ============================================================
print("\n--- Duplicate Rows ---")
for name, df in [("orders", orders_df), ("order_items", order_items_df),
                 ("products", products_df), ("order_reviews", order_reviews_df)]:
    dup_count = df.duplicated().sum()
    print(f"{name}: {dup_count} duplicates")

# ============================================================
# CELL: Assessing Data - Cek Tipe Data
# ============================================================
print("\n--- Data Types: orders_df ---")
print(orders_df.dtypes)
print("\n--- Data Types: order_items_df ---")
print(order_items_df.dtypes)
```

**Steps to Take:**
- Kolom tanggal di orders_df (order_purchase_timestamp, dll.) bertipe object → perlu diubah ke datetime.
- Missing values di orders_df pada kolom order_approved_at, order_delivered_carrier_date, order_delivered_customer_date → perlu ditangani.
- Missing values di products_df pada kolom product_category_name dan dimensi produk → perlu ditangani.
- Missing values di order_reviews_df pada kolom review_comment_title dan review_comment_message → wajar karena opsional, bisa dibiarkan atau diisi "No Comment".
- Duplicate rows di order_reviews_df → perlu dihapus.

**Insight:** (Opsional)
- orders_df memiliki missing values pada kolom delivery dates, kemungkinan karena pesanan belum dikirim/diterima.
- products_df memiliki missing values pada product_category_name (610 baris).
- order_reviews_df memiliki banyak missing values pada kolom komentar (opsional).

### Cleaning Data

#### Konversi Tipe Data Datetime, Handle Missing Values, Hapus Duplikat pada Reviews, Merge tabel untuk analisis

```python
# ============================================================
# CELL: Cleaning Data - Konversi Tipe Data Datetime
# ============================================================
print("=" * 60)
print("CLEANING DATA")
print("=" * 60)

datetime_cols = [
    "order_purchase_timestamp", "order_approved_at",
    "order_delivered_carrier_date", "order_delivered_customer_date",
    "order_estimated_delivery_date"
]
for col in datetime_cols:
    orders_df[col] = pd.to_datetime(orders_df[col], errors='coerce')

print("Tipe data orders_df setelah konversi:")
print(orders_df.dtypes)

# ============================================================
# CELL: Cleaning Data - Handle Missing Values
# ============================================================
# Drop baris orders yang belum delivered (untuk analisis yang fokus pada pesanan selesai)
orders_delivered_df = orders_df[orders_df["order_status"] == "delivered"].copy()
print(f"\nJumlah pesanan delivered: {len(orders_delivered_df)} dari {len(orders_df)} total")

# Handle missing product_category_name
products_df["product_category_name"].fillna("unknown", inplace=True)
print(f"Missing product_category_name setelah cleaning: {products_df['product_category_name'].isnull().sum()}")

# Handle missing review comments (opsional, isi dengan kosong)
order_reviews_df["review_comment_title"].fillna("", inplace=True)
order_reviews_df["review_comment_message"].fillna("", inplace=True)

# ============================================================
# CELL: Cleaning Data - Hapus Duplikat pada Reviews
# ============================================================
before_dup = len(order_reviews_df)
order_reviews_df.drop_duplicates(inplace=True)
after_dup = len(order_reviews_df)
print(f"Review sebelum hapus duplikat: {before_dup}, sesudah: {after_dup}, dihapus: {before_dup - after_dup}")

# ============================================================
# CELL: Cleaning Data - Merge tabel untuk analisis
# ============================================================
# Translate category names to English
products_translated_df = products_df.merge(
    category_translation_df, on="product_category_name", how="left"
)
products_translated_df["product_category_name_english"].fillna(
    products_translated_df["product_category_name"], inplace=True
)

# Merge order_items + products + orders + reviews
analysis_df = (
    order_items_df
    .merge(products_translated_df[["product_id", "product_category_name_english"]], on="product_id", how="left")
    .merge(orders_delivered_df[["order_id", "customer_id", "order_purchase_timestamp"]], on="order_id", how="inner")
    .merge(order_reviews_df[["order_id", "review_score"]], on="order_id", how="left")
)

print(f"\nShape analysis_df: {analysis_df.shape}")
print(analysis_df.head())

# Export main_data.csv untuk dashboard Streamlit
analysis_df.to_csv("dashboard/main_data.csv", index=False)
print("\nmain_data.csv berhasil di-export ke folder dashboard/")
```

**Insight:** (Opsional)
- Data telah dibersihkan: tipe datetime dikonversi, missing values ditangani, duplikat dihapus.
- Dibuat analysis_df dengan menggabungkan order_items, products, orders (delivered), dan reviews.
- Fokus analisis hanya pada pesanan dengan status "delivered".

## Exploratory Data Analysis (EDA)

### Explore Tren Pesanan Bulanan (Pertanyaan 1), Explore Review Score per Kategori (Pertanyaan 2)

```python
# ============================================================
# CELL: EDA - Explore Tren Pesanan Bulanan (Pertanyaan 1)
# ============================================================
print("=" * 60)
print("EXPLORATORY DATA ANALYSIS")
print("=" * 60)

# Hitung jumlah pesanan unik per bulan
orders_delivered_df["order_month"] = orders_delivered_df["order_purchase_timestamp"].dt.to_period("M")
monthly_orders = orders_delivered_df.groupby("order_month")["order_id"].nunique().reset_index()
monthly_orders.columns = ["month", "total_orders"]
monthly_orders["month_dt"] = monthly_orders["month"].dt.to_timestamp()

# Hitung Month-over-Month growth
monthly_orders["mom_growth_pct"] = monthly_orders["total_orders"].pct_change() * 100

print("\n=== Tren Pesanan Bulanan ===")
print(monthly_orders.to_string(index=False))
print(f"\nBulan dengan pesanan tertinggi: {monthly_orders.loc[monthly_orders['total_orders'].idxmax(), 'month']}")
print(f"Jumlah pesanan tertinggi: {monthly_orders['total_orders'].max()}")

# ============================================================
# CELL: EDA - Explore Review Score per Kategori (Pertanyaan 2)
# ============================================================
# Hitung rata-rata review score dan jumlah transaksi per kategori
category_review = analysis_df.groupby("product_category_name_english").agg(
    avg_review_score=("review_score", "mean"),
    total_transactions=("order_id", "count")
).reset_index()

# Filter kategori dengan volume signifikan (>100 transaksi)
category_review_filtered = category_review[category_review["total_transactions"] > 100].copy()
category_review_filtered = category_review_filtered.sort_values("avg_review_score", ascending=True)

print("\n=== Kategori dengan Review Terendah (>100 transaksi) ===")
print(category_review_filtered.head(10).to_string(index=False))

# Kategori dengan review di bawah 4.0
low_review = category_review_filtered[category_review_filtered["avg_review_score"] < 4.0]
print(f"\nJumlah kategori dengan avg review < 4.0: {len(low_review)}")
print(low_review.to_string(index=False))
```

**Insight:** (Opsional)
- Terdapat tren pertumbuhan yang konsisten dari awal 2017 hingga akhir 2017.
- Terjadi puncak pesanan pada bulan tertentu yang perlu divisualisasikan.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Tren Pesanan Bulanan MoM

```python
# ============================================================
# CELL: Visualisasi Pertanyaan 1 - Tren Pesanan Bulanan MoM
# ============================================================
print("\n" + "=" * 60)
print("VISUALIZATION & EXPLANATORY ANALYSIS")
print("=" * 60)

fig, ax1 = plt.subplots(figsize=(14, 7))

# Bar chart untuk jumlah pesanan
color_bars = '#4C72B0'
bars = ax1.bar(monthly_orders["month_dt"], monthly_orders["total_orders"],
               width=25, color=color_bars, alpha=0.7, label="Jumlah Pesanan", edgecolor='white')

ax1.set_xlabel("Bulan", fontsize=13, fontweight='bold')
ax1.set_ylabel("Jumlah Pesanan", fontsize=13, fontweight='bold', color=color_bars)
ax1.tick_params(axis='y', labelcolor=color_bars)
ax1.tick_params(axis='x', rotation=45)

# Highlight bulan puncak
peak_idx = monthly_orders["total_orders"].idxmax()
bars[peak_idx].set_color('#E74C3C')
bars[peak_idx].set_alpha(1.0)

# Line chart untuk MoM Growth di axis kedua
ax2 = ax1.twinx()
color_line = '#E67E22'
ax2.plot(monthly_orders["month_dt"], monthly_orders["mom_growth_pct"],
         color=color_line, marker='o', linewidth=2.5, markersize=6, label="MoM Growth (%)")
ax2.set_ylabel("MoM Growth (%)", fontsize=13, fontweight='bold', color=color_line)
ax2.tick_params(axis='y', labelcolor=color_line)
ax2.axhline(y=0, color='gray', linestyle='--', alpha=0.5)

plt.title("Tren Jumlah Pesanan Bulanan & Pertumbuhan MoM\nE-Commerce Olist (2017-2018)",
          fontsize=15, fontweight='bold', pad=20)

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left', fontsize=11)

plt.tight_layout()
plt.savefig("viz_pertanyaan1_tren_pesanan.png", dpi=150, bbox_inches='tight')
plt.show()
```

## Analisis Lanjutan (Opsional)

```python
# ============================================================
# ANALISIS LANJUTAN (Opsional) - RFM Analysis Sederhana
# ============================================================
print("\n" + "=" * 60)
print("ANALISIS LANJUTAN - RFM Sederhana")
print("=" * 60)

# Merge orders dengan customers
rfm_base = orders_delivered_df.merge(customers_df[["customer_id", "customer_unique_id"]], on="customer_id")
rfm_base = rfm_base.merge(
    order_payments_df.groupby("order_id")["payment_value"].sum().reset_index(),
    on="order_id"
)

reference_date = rfm_base["order_purchase_timestamp"].max() + pd.Timedelta(days=1)

rfm_df = rfm_base.groupby("customer_unique_id").agg(
    recency=("order_purchase_timestamp", lambda x: (reference_date - x.max()).days),
    frequency=("order_id", "nunique"),
    monetary=("payment_value", "sum")
).reset_index()

print(rfm_df.describe())

# ============================================================
# CELL: Visualisasi RFM Analysis
# ============================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Recency
axes[0].hist(rfm_df["recency"], bins=50, color="#4C72B0", edgecolor="white", alpha=0.8)
axes[0].set_title("Distribusi Recency (Hari)", fontsize=13, fontweight="bold")
axes[0].set_xlabel("Recency (hari sejak pembelian terakhir)")
axes[0].set_ylabel("Jumlah Pelanggan")

# Frequency
axes[1].hist(rfm_df["frequency"], bins=range(1, rfm_df["frequency"].max() + 2),
             color="#E67E22", edgecolor="white", alpha=0.8)
axes[1].set_title("Distribusi Frequency", fontsize=13, fontweight="bold")
axes[1].set_xlabel("Jumlah Transaksi")
axes[1].set_ylabel("Jumlah Pelanggan")

# Monetary
axes[2].hist(rfm_df["monetary"], bins=50, color="#27AE60", edgecolor="white", alpha=0.8)
axes[2].set_title("Distribusi Monetary (R$)", fontsize=13, fontweight="bold")
axes[2].set_xlabel("Total Pengeluaran (R$)")
axes[2].set_ylabel("Jumlah Pelanggan")

plt.suptitle("RFM Analysis - Segmentasi Pelanggan E-Commerce Olist",
             fontsize=15, fontweight="bold", y=1.02)
plt.tight_layout()
plt.savefig("viz_rfm_analysis.png", dpi=150, bbox_inches="tight")
plt.show()

# **Insight RFM:**
# - Mayoritas pelanggan hanya melakukan 1 transaksi (frequency=1), menunjukkan rendahnya repeat purchase.
# - Distribusi monetary menunjukkan sebagian besar pelanggan berbelanja dalam jumlah kecil-menengah.
# - Pelanggan dengan recency rendah (baru belanja) dan monetary tinggi adalah target utama program loyalitas.
```
